```python
import pickle
import numpy as np
import pandas as pd
import numpy as np
```

```python
train_file = "data/fii-nn-2025-homework-2/extended_mnist_train.pkl"
test_file = "data/fii-nn-2025-homework-2/extended_mnist_test.pkl"

with open(train_file, "rb") as fp:
    train = pickle.load(fp)

with open(test_file, "rb") as fp:
    test = pickle.load(fp)
```

```python
train_data = []
train_labels = []
for image, label in train:
    train_data.append(image.flatten() / 255.0)
    train_labels.append(label)
```

```python
test_data = []
for image, label in test:
    test_data.append(image.flatten() / 255.0)
```

```python
#exagerates large numbers and after that it normalize
def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)
```

```python
X = np.array(train_data)
Y = np.array(train_labels)

n_inputs = 784
n_outputs = 10
n_epochs = 100
W = np.zeros((n_inputs, n_outputs))
B = np.zeros(n_outputs)
batch_size = 32
lr = 0.01

Y_binary = np.zeros((Y.shape[0], n_outputs))
for i in range(n_outputs):
    Y_binary[:, i] = np.where(Y == i, 1, 0)

for epoch in range(n_epochs):
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    X_shuffled = X[indices]
    Y_shuffled = Y_binary[indices]

    for i in range(0, X.shape[0], batch_size):
        X_batch =X_shuffled[i : i + batch_size]
        Y_batch = Y_shuffled[i : i + batch_size]
        dim_batch = X_batch.shape[0]

        z = np.dot(X_batch, W) + B
        z = softmax(z)

        dz = z - Y_batch
        dw = np.dot(X_batch.T, dz) / dim_batch
        db = np.sum(dz, axis=0) / dim_batch

        W -= lr * dw
        B -= lr * db

        if epoch % 10 == 0 or epoch == n_epochs - 1:
            loss = -np.sum(Y_batch * np.log(z)) / dim_batch
            # print(f"Epoch {epoch}, Loss: {loss:.4f}")


```

```python
def predict(X, W, B):
    z = np.dot(X, W) + B
    z = softmax(z)
    return np.argmax(z, axis=1)
```

```python
predictions = predict(test_data, W, B)
```

```python
    # This is how you prepare a submission for the competition
predictions_csv = {
    "ID": [],
    "target": [],
}

for i, label in enumerate(predictions):
    predictions_csv["ID"].append(i)
    predictions_csv["target"].append(label)

df = pd.DataFrame(predictions_csv)
df.to_csv("submission.csv", index=False)
```
